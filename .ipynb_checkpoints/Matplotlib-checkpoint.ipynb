# 2D Plot
A 2D plot (or 2-dimensional plot) is a type of graph where:

- One variable is plotted along the X-axis (horizontal axis)
- Another variable is plotted along the Y-axis (vertical axis)


### 2D line Plot Used in-
- **Bivariate Analysis** -Bivariate analysis is a type of statistical analysis that looks at the relationship between two different variables.
1. "Bi" means two
2. "Variate" refers to variables
- **Can be applied on** - categorical -> numerical /numerical -> numerical
- **Use case** - Time series data (Time series data is a sequence of data points collected or recorded at specific time intervals.)
eg. increasing covid cases day by day 

### plt.plot()
used to draw 2D line plots. It connects your data points with lines by default.

### plotting a simple function

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price) 
```

###  plotting from a pandas dataframe

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

### plotting multiple plots

```python
# First line
plt.plot(batsman['index'],batsman['V Kohli'])
# Second line
plt.plot(batsman['index'],batsman['RG Sharma'])
```

### Customizing Plots
Matplotlib allows fine-grained control:

- **Line style**: solid, dashed, dotted
- **Colors**: using color names, hex codes, RGB tuples
- **Markers**: dots, squares, stars, etc.
- **Grid lines**
- **Legends**
- **Multiple plots in one figure** (subplotting)

### Adding Tittle and Label to the graph

```python
# labels, title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison') # adds tittle to the graph
plt.xlabel('Season') # adds label on x axis
plt.ylabel('Runs Scored')   #adds label on y axis
```

### Changing colour


```python
# change color
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### Changing Line color and Line width 
You can do this using the color and linewidth parameters inside plt.plot().

```python
# Line plot with custom color and line width
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### Manipulating markers and Controlling marker size

**Common Marker Styles:**


```python
# manipulating markers and controlling marker size
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### Adding Legend
**Step 1**: Use the label parameter inside plt.plot()    
**Step 2**: Call plt.legend() to display it

```python
# adding legend

# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()   #plt.legend(loc='upper right')
```

### Limiting axes
limiting the axes in Matplotlib is super handy when you want to control what part of the data is visible on your plot or to zoom into a specific range
- **plt.xlim()** â†’ to set X-axis limits
- **plt.ylim()** â†’ to set Y-axis limits



```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)   # set limits for the y-axis
plt.xlim(2017,2019) # set limits for the x-axis
```

```python
# grid --> adds grid to the graph 
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show ---> tells Python to display the current figure(s) on the screen.
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

# Scatter Plots
A scatter plot displays points (markers) representing two sets of values.</br>
Itâ€™s great for showing the relationship (correlation, clustering, or spread) between two continuous variables.


## Scatter Plots used in 
- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

### plt.scatter()
plt.scatter() is a function that creates a scatter plot â€” a type of plot that displays points (as markers) at specific (x, y) positions on a 2D graph.


### Plotting simple function

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

### scatter Plot on pandas data

```python
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='.')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

### Controlling marker size
You use the s parameter to control the size of each marker.
Important:
- The value of s represents the area of the marker in pointsÂ² (not the radius or diameter).
- Bigger values mean bigger markers.

```python
# size
tips = sns.load_dataset('tips')  # dataset in seaborn
tips
```

```python
# Example: Same Size for All Markers
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# s=100 sets the area of each marker
plt.scatter(x, y, s=100, color='blue')

plt.title('Scatter Plot with Fixed Marker Size')
plt.grid(True)
plt.show()
```

```python
# slower technique
# Example: Different Sizes for Each Marker
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# scatterplot using plt.plot
# faster technique to plot scatter graph
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# Bar chart
A bar chart displays data using rectangular bars, where the height (or length) of each bar represents the value of a category.


## Bar chart used in 
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

## plt.bar()
Used to plot barchart in matplotlib

## Simple bar chart

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children) # plt.bar() --> used to plot bar chart
```

### horizontal bar chart

```python
# If number of categories are increased then use horizontal bar chart
plt.barh(colors,children,color='black')
```

### Multiple bar chart

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# Customizing Bar Colors & Width in Multiple barchart
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem----> if the name of the xlabel is too long
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical') # set rotation
plt.show()
```

```python
# Stacked Bar chart
# Plot the first set (bottom of the stack)
plt.bar(df['batsman'],df['2017'],label='2017')
# Plot the second set (on top of the first)
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
# Plot the third set (on top of the second)
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

# Histogram
A histogram is a type of bar chart that groups numeric data into bins (interval ranges) and shows how many data points fall into each bin.


## Histogram is used in 
- Univariate Analysis- used for only one column
- Numerical col
- Use case - Frequency Count

### Histogram on Simple data

```python
# simple data

data = [32,45,56,10,15,27,61]

plt.hist(data)   #plt.hist() used to plot histogram
```

### Handling bins


```python
# handling bins
data = [32,45,56,10,15,27,61]

plt.hist(data,bins=[10,25,40,55,70])
```

# Pie Chart
A pie chart is a circular chart divided into slices to show proportions of a whole.
Each slice represents a categoryâ€™s contribution to the total.


## Used on
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

### Piechart on simple data

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
# add lables
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

# Piechart on Pandas data

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

### Adding percent 
You can use the autopct parameter inside plt.pie() to display percentage values directly on the slices.

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%') #Format the number with 1 decimal place
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

## Explode 
The explode parameter allows you to "pull out" one or more slices of the pie chart for emphasis. You can create a visual effect where a slice is separated slightly from the rest of the chart.
## shadow=True â†’ 
Adds a shadow effect to the whole pie chart.

```python
# explode and shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
