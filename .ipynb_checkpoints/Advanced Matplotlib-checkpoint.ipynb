```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplots


```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# same Color
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# colour encoding on the basis of species
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
#Add color
# Using cmap and alpha (alpha controls transparency, its value lies between 0 and 1)
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Plot size
Use the figsize argument inside plt.figure()
The size is in inches (not pixels)  one inch roughly equals 100 pixels on most screens.


```python
# Increase size of the graph 
plt.figure(figsize=(10,5))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotations
Annotations let you add text labels to specific points or positions in your plot.


```python
batters = pd.read_csv('batter.csv')
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
# example
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'})
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Horizontal and Vertical lines



```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Subplots
subplots in Matplotlib are used when you want to show multiple plots in one figure  side by side, stacked, or in a grid.

```python
batters.head()
```

```python
# First way to plot graph
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# A diff way to plot graphs
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

```


_____________________________________________________________________


```python
# plotting two graphs 
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# four graphs
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
# Third way to plot graphs 
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)   # means 2 rows, 2 columns, and you're referencing the first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)  # means  2 rows, 2 columns and you're referencing the second plot
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)  # means  2 rows, 2 columns and you're referencing the third plot
ax3.hist(batters['avg'])
```

# 3D Scatter Plots

```python
fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

# 3D Surface Plots
A 3D surface plot in matplotlib is used to represent three-dimensional data on a continuous surface. It's especially useful when you have a function of two variables (e.g., 

(

,

)
f(x,y)) and you want to visualize it in 3D.

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Contour Plots
A contour plot is a way to represent 3D data in 2D

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Filled Contour Plot
Used if you want the areas between the lines to be colored.

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Heatmap
A heatmap is a 2D visualization where data values are represented by colors. Its super useful when you want to see patterns, correlations, or intensities in a matrix or grid-like dataset.<br>
When to Use a Heatmap:
- To visualize correlation matrices.
- To display a 2D array of values (e.g., pixel intensities, confusion matrices).
- To spot trends or clusters in data.
- To compare two categorical values side by side.

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(5,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```

# Pandas Plot()
plot() is a function present in pandas using which you can plot matplotlib's graph inside pandas. <br>
Pandas provides a .plot() method on both Series and DataFrames that wraps around matplotlib functions.

### **On series**

```python
s = pd.Series([1,2,3,4,5,6,7])
s.plot(kind='line')
```

```python
s = pd.Series([1,2,3,4,5,6,7])
s.plot(kind='pie')
```

### **On Dataframe**

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.plot(kind='scatter', x='total_bill', y='tip')
```

```python
tips['size'] = tips['size'] * 100
```

```python
# adding tittle, changing marker, figsize and color
# Scatter plot -> labels -> markers -> figsize -> color -> cmap
tips.plot(kind='scatter',x='total_bill',y='tip',title='Cost Analysis',marker='+',figsize=(10,6),s='size',c='sex',cmap='viridis')
```

## Plotting Pie chart

**1. single column**

```python
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
# single column
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%')
```

**2. Multiple column**

```python
# multiple pie charts

df[['match1','match2','match3']].plot(kind='pie',subplots=True,figsize=(15,8))
```
